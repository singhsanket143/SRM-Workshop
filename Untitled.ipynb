```python
def load_text(filename):
    with open(filename, encoding="utf-8") as f:
        return f.read().lower()
```

```python
data = load_text("data1.txt")
print(data)
```

```python
def preprocess(string, k):
    T = {}
    for i in range(len(string) - k):
        X = string[i:i+k]
        Y = string[i+k]
        #print(X, Y)
        if T.get(X) is None:
            T[X] = {}
            T[X][Y] = 1
        elif T[X].get(Y) is None:
            T[X][Y] = 1
        else:
            T[X][Y] += 1
        
    return T
```

```python
preprocess_data = preprocess(data, 4)
print(preprocess_data)
```

```python
def changeToProbability(T):
    for kx in T.keys():
        norm_factor = sum(T[kx].values())
        for val in T[kx].keys():
            T[kx][val] = T[kx][val] / norm_factor
    
    return T
```

```python
final_dataset = changeToProbability(preprocess_data)
print(final_dataset)
```

```python
import numpy as np
```

```python
def predict(T, string):
    candidate = list(T[string].keys())
    probab = list(T[string].values())
    return np.random.choice(candidate, p=probab)
```

```python
sentence = "कौन "
k = 4
for i in range(1000):
    last = sentence[-k:]
    new_char = predict(final_dataset, last)
    sentence += new_char
    
print(sentence)
```
